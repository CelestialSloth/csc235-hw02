```python
import plotly.express as px, pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
```

```python
comm_sun_df = pd.read_csv('hw02-data/Communication Data/comm-data-Sun.csv')
```

```python
comm_sun_df['Timestamp'] = pd.to_datetime(comm_sun_df['Timestamp'])
comm_sun_df_filtered = comm_sun_df.query('location=="Wet Land" & (Timestamp>"2014-06-08T11:30:00") & Timestamp<"2014-06-08T12:30:00"').copy()
# comm_sun_df_filtered[comm_sun_df_filtered['to'] != '839736']['to'] = 
# comm_df_filter = pd.eval('to~=839736 & to~=1278894 & to~=external')
# print((comm_sun_df_filtered['to']=='839736').dtype)
# comm_sun_df_filtered[(comm_sun_df_filtered['to']=='839736') | (comm_sun_df_filtered['to']=='1278894') | (comm_sun_df_filtered['to']=='external')]
comm_sun_df_filtered.loc[~((comm_sun_df_filtered['to']=='839736') | (comm_sun_df_filtered['to']=='1278894') | (comm_sun_df_filtered['to']=='external')),'to'] = "other"
```

```python
fig = px.pie(comm_sun_df_filtered, names='to', title='Top Receivers of Messages from Wet Land, Sunday 11:30am-12:30pm ')
fig.show()
```

```python
moves_sun_df  = pd.read_csv('hw02-data/Movement Data/park-movement-Sun.csv', skiprows=[4332993,10932426,10932427],dtype={'Timestamp': object, 'id': int, 'type': str, 'X': int, 'Y': int})
```

```python
moves_sun_df['Timestamp'] = pd.to_datetime(moves_sun_df['Timestamp'])
```

```python
moves_sun_df_filtered_pavilion = moves_sun_df.query('X==32 & Y==33 & type=="check-in"').copy()
moves_sun_df_filtered_pavilion['ten_min'] = moves_sun_df_filtered_pavilion['Timestamp'].dt.round('5min')
frequency_moves_sun_df_filtered_pavilion = moves_sun_df_filtered_pavilion.groupby('ten_min').size().reset_index(name='Calls')
```

```python
fig = go.Figure(go.Scatter(x=frequency_moves_sun_df_filtered_pavilion['ten_min'], 
               y=frequency_moves_sun_df_filtered_pavilion['Calls'], 
               mode='lines', name='Sunday', 
               fill='tozeroy',))
fig.show()
```
