```python
import plotly.graph_objects as go
import pandas as pd
```

```python
#Sunday df with amount of calls per 5min intervals
comm_sun_df = pd.read_csv('hw02-data/Communication_data/comm-data-Sun.csv')
comm_sun_df['Timestamp'] = pd.to_datetime(comm_sun_df['Timestamp'])
comm_sun_df['one_min'] = comm_sun_df['Timestamp'].dt.round('min')
frequency_comm_sun_df = comm_sun_df.groupby('one_min').size().reset_index(name='calls')
frequency_comm_sun_df.head()
```

```python
# Create the main figure (focus area)
fig = go.Figure()

# Plot the area chart for the 'focus' (main zoomed-in plot)
fig.add_trace(
    go.Scatter(
        x=frequency_comm_sun_df['one_min'],
        y=frequency_comm_sun_df['calls'],
        mode='lines',
        fill='tozeroy',  # fill the area under the line
        name='Price'
    )
)

# Add the range slider and range selector to simulate the 'brush' functionality
fig.update_layout(
    title="Call Density Per Minute on Sunday",
    yaxis=dict(title='Calls'),
    xaxis=dict(
        title='Time Range Selection for Sunday',
        rangeslider=dict(visible=True),  # Enables the range slider (simulates brushing)
        rangeselector=dict(  # Adds quick range selection buttons
            buttons=list([
                dict(step="all")
            ])
        )
    )
)
```
